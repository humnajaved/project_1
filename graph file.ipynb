```python
import numpy as np
import matplotlib.pyplot as plt

minerals = ['Quartz', 'Calcite', 'Dolomite', 'Illite', 'Kaolinite', 'Pyrite', 'K-Feldspar']

initial_weights_eagle_ford = [30, 10, 20, 5, 15, 7, 13]
experimental_eagle_ford = [28, 9, 21, 6, 14, 8, 12]
phreeqc_model_eagle_ford = [29, 11, 19, 6, 16, 7, 13]
variation_ef = [5.7, 3.5, 4.0, 5.3, 5.4, 2.5, 2.9]  # Variations 

initial_weights_mancos = [35, 12, 18, 4, 12, 9, 10]
experimental_mancos = [33, 11, 19, 5, 13, 8, 11]
phreeqc_model_mancos = [34, 13, 17, 5, 14, 9, 10]
variation_mancos = [5.5, 5.4, 6.4, 5.3, 4.7, 3.5, 2.9]  # Variations

def plot_mineral_comparison(minerals, initial, experimental, model, variation, title):
    bar_width = 0.25
    index = np.arange(len(minerals))

    plt.figure(figsize=(10, 6))
    
    # Bars for Initial weights
    plt.bar(index, initial, bar_width, label='Initial Weight %', color='orange')
    
    # Bars for Experimental results
    plt.bar(index + bar_width, experimental, bar_width, label='Experimental', color='yellow')
    
    # Bars for PHREEQC Model results
    plt.bar(index + 2 * bar_width, model, bar_width, label='PHREEQC Model', color='red')

    plt.xlabel('Minerals')
    plt.ylabel('Mass Fraction (%)')
    plt.title(title)
    plt.xticks(index + bar_width, minerals)
    
    # Plotting variations as dots
    plt.scatter(index + bar_width * 1.5, variation, color='black', label='Variation %')

    plt.legend()
    plt.tight_layout()
    plt.show()

# Plot for Eagle Ford
plot_mineral_comparison(minerals, initial_weights_eagle_ford, experimental_eagle_ford, phreeqc_model_eagle_ford, variation_ef, 'Eagle Ford')

# Plot for Mancos
plot_mineral_comparison(minerals, initial_weights_mancos, experimental_mancos, phreeqc_model_mancos, variation_mancos, 'Mancos')
```
